```python
import pandas as pd
import numpy as np
```

## Load the Dataset
The dataset is loaded from the raw data folder for inspection and cleaning.

```python
df = pd.read_csv("../data/raw/onlinefraud.csv")
df.head()
```

## Check for Missing Values
I check each column to Identify columns that may require treatment.

```python
df.isnull().sum()
```

### Findings
No missing values were found in any column. Does not require missing value treatment.

## Check for Duplicate Records
I check for duplicate rows to ensure that each transaction is represented only once in the dataset.

```python
df.duplicated().sum()
```

### Findings
No duplicate records were found in the dataset.
This indicates that each transaction is unique and no duplicate removal was required before analysis.


## Transaction Types
I examine the trancsaction categories to identify misspellings, unexpected categories or inconsistent labels.

```python
df["type"].unique()
```

```python
df["type"].value_counts()
```

### Findings
No inconsistencies or spelling variations were observed in the transaction categories, indicating that the data is clean and standardized. 

## Fraud Distribution
The target variable indicates whether a transaction is fraudulent or legitimate.

```python
df["isFraud"].value_counts()
```

```python
df["isFraud"].value_counts(normalize=True) * 100
```

### Findings
The dataset is highly imbalanced.
Out of 6,362,620 transactions, only 8,213 transactions were identified as fraudulent, representing approximately 0.13%.
In contrast, legitimate transactions account for approximately 99.87% of the dataset.
Fraudulent activities are rare compared to normal financial transactions.

## Flagged Fraud Transactions
This variable indicates whether a transaction was automatically flagged as suspicious by the system.
I examine its distribution to understand how frequently transactions were flagged.

```python
df["isFlaggedFraud"].value_counts()
```

```python
df["isFlaggedFraud"].value_counts(normalize=True) * 100
```

### Findings
Only 16 transactions were automatically flagged as suspicious by the system, representing less than 0.001% of all transactions.

## Check for Zero-Value Transactions
I examine the dataset to determine whether any transactions were recorded without a monetary value..

```python
(df["amount"] == 0).sum()
```

### Findings
A total of 16 transactions were recorded with a transaction amount equal to zero
These records will be retained for further investigation during EDA, as they may contain useful information related to fraud detection or system-generated transactions.

## Data Cleaning Summary

The dataset passed all major data quality checks.

No missing values, duplicate records, invalid data types, or negative monetary values were identified.

The dataset is considered clean and suitable for exploratory data analysis.
